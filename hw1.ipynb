```python
import os

INPUT = '/content/drive/MyDrive/news_vnexpress'

print("Các nhãn và số văn bản tương ứng trong dữ liệu:")
n = 0

for label in os.listdir(INPUT):
    label_path = os.path.join(INPUT, label)

    if os.path.isdir(label_path):
        count = len(os.listdir(label_path))
        print(f'{label}: {count}')
        n += count
    else:
        print(f' -> [Đã bỏ qua file rác]: {label}')

print(f"----------------------------------")
print(f"Tổng số văn bản hợp lệ: {n}")
```

```python
data_train = load_files(container_path=INPUT, encoding="utf-8")
print('mapping:')
for i in range(len(data_train.target_names)):
    print(f'{data_train.target_names[i]} - {i}')
print(data_train.filenames[0:1])
print(data_train.target[0:1])
print(data_train.data[0:1])
print("\nTong so van ban : {}" .format(len(data_train.filenames)))
```

```python
import requests
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline

url = "https://raw.githubusercontent.com/stopwords/vietnamese-stopwords/master/vietnamese-stopwords.txt"
response = requests.get(url)

stopwords_raw = response.text.split('\n')

stopwords = [x.strip().replace(" ", "_") for x in stopwords_raw if x.strip()]

print(f"So luong stopwords: {len(stopwords)}")
print(stopwords[:10])
module_count_vector = CountVectorizer(stop_words=stopwords)
model_rf_preprocess = Pipeline([
    ('vect', module_count_vector),
    ('tfidf', TfidfTransformer()),
])

data_preprocessed = model_rf_preprocess.fit_transform(data_train.data, data_train.target)

X = data_preprocessed
Y = data_train.target

print(f"\nSo luong tu trong tu dien: {len(module_count_vector.vocabulary_)}")
print(f"Kich thuoc du lieu sau khi xu ly: {X.shape}")
print(f"Kich thuoc nhan tuong ung : {Y.shape}")
```
